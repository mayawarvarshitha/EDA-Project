## 7. Find Different Payment Methods Used By Customers Citywise. 

```python
# Creating a crosstab
crosstab_data = pd.crosstab(df['City'], df['Payment'])

# Creating a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab_data, annot=True, fmt='d', cmap='Blues')

plt.xlabel('Payment')
plt.ylabel('City')
plt.title('Crosstab Heatmap of Payment by City')
plt.show()
```

## 9. Display Daily Sales by Day of the Week

```python
dw_mapping= {
    0:'Mon',
    1:'Tue',
    2:'Wed',
    3:'Thur',
    4:'Fri',
    5:'Sat',
    6:'Sun'
}
```

```python
df['day_of_week'] =df['Date'].dt.dayofweek.map(dw_mapping)
df.head()
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
```

```python
'''From this bar plot we can see that Sales is higher in Saturday because its weekeng
, by this i can Manage my Staffing and Shift timing
Suprisingly Sales are high on Tuesday,
Sales are Lowest on Monday by this i can Keep Offers in lowest Sales Days to Get Better Sales'''
```

## 10. What Will Be The Highest Months for Sales

```python
month_mapping = {
    1:'Jan',
    2:'Feb',
    3:'March'
    
}
```

```python
df['month'] =df['Date'].dt.month.map(month_mapping)
df.head()
```

```python
df['month'].value_counts().plot(kind='bar')
```
